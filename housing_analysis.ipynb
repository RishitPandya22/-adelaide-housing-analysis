```python
# Create price category labels
def price_category(price):
    if price < 500000:
        return 'Budget'
    elif price < 800000:
        return 'Mid-Range'
    else:
        return 'Premium'

df['Price_Category'] = df['Median_Price'].apply(price_category)

# Check it worked
df[['Suburb', 'Median_Price', 'Price_Category']].head(10)
```

```python
# Make numbers readable
pd.set_option('display.float_format', lambda x: f'${x:,.0f}')

# Now rerun Question 1
suburb_prices = df.groupby('Suburb')['Median_Price'].mean().sort_values(ascending=False)
print(suburb_prices)
```

```python
## Finding 1: Most Expensive Suburbs
# Burnside and Unley are Adelaide's most expensive suburbs with median prices 
# above $1M. Elizabeth and Salisbury are the most affordable, reflecting their 
# distance from the CBD and demographic differences.
```

```python
# Average price per year across all suburbs
yearly_prices = df.groupby('Year')['Median_Price'].mean()
print(yearly_prices)
```

```python
# ## Finding 2: Year Over Year Price Trend
# Adelaide housing prices have grown consistently from 2019 to 2022, with 
# average prices rising significantly each year. This reflects the broader 
# Australian property boom during this period.
```

```python
# Average price grouped by distance to CBD
distance_price = df.groupby('Distance_to_CBD_km')['Median_Price'].mean().sort_values(ascending=False)
print(distance_price)
```

```python
# ## Finding 3: Distance vs Price
# There is a clear negative relationship between distance to CBD and median 
# price — suburbs closer to the city command significantly higher prices. 
# This confirms that location remains the primary driver of property value.
```

```python
# Count of suburbs in each price category
category_counts = df['Price_Category'].value_counts()
print(category_counts)
```

```python
# ## Finding 4: Market Distribution
# The Adelaide housing market is fairly balanced between Budget, Mid-Range 
# and Premium properties, with Mid-Range properties being most common. 
# This suggests Adelaide remains more affordable than Sydney or Melbourne.
```

```python
# Price in 2019
prices_2019 = df[df['Year'] == 2019].groupby('Suburb')['Median_Price'].mean()

# Price in 2022
prices_2022 = df[df['Year'] == 2022].groupby('Suburb')['Median_Price'].mean()

# Growth calculation
price_growth = ((prices_2022 - prices_2019) / prices_2019 * 100).sort_values(ascending=False)
print("Price Growth % from 2019 to 2022:")
print(price_growth.round(2))
```

```python
# ## Finding 5: Fastest Growing Suburbs
# Inner city suburbs like Norwood and Burnside have seen the highest price 
# growth from 2019 to 2022, making them strong investment areas. Outer 
# suburbs showed slower but steady growth.
```

```python
# Create a map centered on Adelaide CBD
adelaide_map = folium.Map(
    location=[-34.9285, 138.6007],  # Adelaide CBD coordinates
    zoom_start=11,
    tiles='CartoDB positron'        # Clean light theme
)

print("Base map created! ✅")
```

```python
# Colour based on price category
def get_color(price_category):
    if price_category == 'Premium':
        return 'red'
    elif price_category == 'Mid-Range':
        return 'orange'
    else:
        return 'green'
```

```python
# Use most recent year data only (2022)
df_2022 = df[df['Year'] == 2022]

# Add a marker for each suburb
for _, row in df_2022.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=12,
        color=get_color(row['Price_Category']),
        fill=True,
        fill_color=get_color(row['Price_Category']),
        fill_opacity=0.7,
        popup=folium.Popup(
            f"""
            <b>{row['Suburb']}</b><br>
            💰 Median Price: ${row['Median_Price']:,.0f}<br>
            🛏️ Bedrooms: {row['Bedrooms']}<br>
            📍 Distance to CBD: {row['Distance_to_CBD_km']} km<br>
            🏷️ Category: {row['Price_Category']}
            """,
            max_width=200
        ),
        tooltip=row['Suburb']
    ).add_to(adelaide_map)

print("Markers added! ✅")
```

```python
# Add legend to map
legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; z-index: 1000;
     background-color: white; padding: 15px; border-radius: 10px;
     border: 2px solid grey; font-size: 14px;">
     <b>🏠 Adelaide Housing Prices</b><br><br>
     <span style="color:red;">●</span> Premium (>$800K)<br>
     <span style="color:orange;">●</span> Mid-Range ($500K-$800K)<br>
     <span style="color:green;">●</span> Budget (<$500K)
</div>
"""
adelaide_map.get_root().html.add_child(folium.Element(legend_html))

print("Legend added! ✅")
```

```python
# Save as interactive HTML file
adelaide_map.save('adelaide_housing_map.html')
print("Map saved! ✅")
```
